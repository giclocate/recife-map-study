### Adicionando marcadores

```python
mapa = folium.Map([-8.063145725766562, -34.87106202258222], zoom_start=12)

folium.Marker(
    location=[-8.063145725766562, -34.87106202258222],
    tooltip="Clique aqui!",
    popup="Praça do Marco Zero",
    icon=folium.Icon(icon="cloud"),
).add_to(mapa)

folium.Marker(
    location=[-8.063109828093529, -34.87392590465414],
    tooltip="Clique aqui!",
    popup="Porto Digital",
    icon=folium.Icon(color="green"),
).add_to(mapa)

mapa
```

### Adicionando vetores de linha

```python
mapa = folium.Map(location=[-8.063111181604691, -34.8738311097181], zoom_start=18)

trail_coordinates = [
    (-8.063111181604691, -34.8738311097181),
    (-8.06127874921586, -34.87151904558318),
    (-8.058248745907285, -34.872016389299795),
    (-8.058815865631292, -34.8698808084941457),
    (-8.063074871814273, -34.87115532917372),
]

folium.PolyLine(trail_coordinates, tooltip="Coast").add_to(mapa)

mapa
```

### Adicionando círculo

```python
mapa = folium.Map(location=[-8.063111181604691, -34.8738311097181], zoom_start=18)

folium.CircleMarker(
    [-8.063111181604691, -34.8738311097181],
    radius = 50,
    color = '#3186cc',
    fill = True,
    fill_collor = '#3186cc'
).add_to(mapa)

mapa
```

### Adicionando pontos no mapa a partir de um dataframe

```python
import pandas as pd
df = pd.read_excel('pontos_recife.xlsx')
df.head()
```

```python
mapa = folium.Map(location = [-8.063111181604691, -34.8738311097181])
for local, lat, lon in zip(df.LOCAL, df.LATITUDE.values, df.LONGITUDE.values):
    folium.Marker([lat, lon], tooltip=f"{local}").add_to(mapa)

mapa
```
